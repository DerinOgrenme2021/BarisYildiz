Ev fiyatının veri kümesinde median_house_value değeri 200000 veya üzerinde ve population değeri 1500 altında olan evlerin histogram garifği

```python
housing_value = cf_housing_df[cf_housing_df['median_house_value'] >= 200000 ] 
housing_value              
```

```python
housing_value_population = housing_value[housing_value['population'] < 1500]
housing_value_population
```

```python
housing_value_population.describe()
```

```python
housing_value_population.hist('median_house_value')
```

```python
housing_value_population_agg_min = housing_value_population.groupby('households').agg(['min'])
housing_value_population_agg_min
```

```python
housing_value_population_agg_min.hist('median_house_value')
```
